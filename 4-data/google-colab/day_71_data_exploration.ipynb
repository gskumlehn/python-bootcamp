```python
spread_col = clean_df["Mid-Career 90th Percentile Salary"] - clean_df["Mid-Career 10th Percentile Salary"]
spread_col
```

```python
clean_df.insert(1, "Spread", spread_col)
```

```python
clean_df.head()
```

```python
low_risk = clean_df.sort_values("Spread")
low_risk[["Undergraduate Major","Spread"]].head()
```

Today's Learning Points





1.   Use .head(), .tail(), .shape and .columns to explore your 
2.   DataFrame and find out the number of rows and columns as well as the column names.
3.   Look for NaN (not a number) values with .findna() and consider using .dropna() to clean up your DataFrame.
4.   You can access entire columns of a DataFrame using the square bracket notation: df['column name'] or df[['column name 1', 'column name 2', 'column name 3']]
5.   You can access individual cells in a DataFrame by chaining square brackets df['column name'][index] or using df['column name'].loc[index]
6.   The largest and smallest values, as well as their positions, can be found with methods like .max(), .min(), .idxmax() and .idxmin()
7.   You can sort the DataFrame with .sort_values() and add new columns with .insert()
8.   To create an Excel Style Pivot Table by grouping entries that belong to a particular category use the .groupby() method
